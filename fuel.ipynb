```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
import pickle
```

```python
data = pd.read_csv('Fuel Efficiency.csv')
```

```python
data
```

```python
data = data.dropna()
```

```python
# Encode categorical columns using LabelEncoder
label_encoders = {}
categorical_columns = ['Transmission', 'Vehicle class', 'Fuel type']
for column in categorical_columns:
    label_encoders[column] = LabelEncoder()
    data[column] = label_encoders[column].fit_transform(data[column])
```

```python
with open('label_encoders.pkl','wb') as f:
    pickle.dump(label_encoders,f)
```

```python
# Define features and target
features = ["Vehicle class", "Engine size (L)", "Cylinders",
            "Transmission", "Fuel type", "CO2 rating"]
X = data[features]
y = data["Combined (L/100 km)"]
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize and fit Random Forest Regression model
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error:', mse)
print('R^2 Score:', r2)
```

```python
# Save the model as a pickle file
with open('rf_model.pkl', 'wb') as f:
    pickle.dump(rf_model, f)
```
